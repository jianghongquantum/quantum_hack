# 电池-风电联合调度量子化（近似路线 + 约束/哈密顿量 + 比特预算）中文笔记

> 原问题（24小时、风电给定、价格给定）  
> 最大化收益：
> $$
> \max \ \sum_{t=1}^{24} p(t)\,[w(t)-c(t)+d(t)]
> $$
> 约束：
> $$
> soc(t)=soc(t-1)+\eta\,c(t)-d(t),\quad \eta=0.8
> $$
> $$
> 0\le c(t)\le 5(1-y(t)),\ y(t)\in\{0,1\}
> $$
> $$
> 0\le c(t)\le w(t)
> $$
> $$
> 0\le d(t)\le 4y(t),\quad 0\le d(t)\le soc(t-1)
> $$
> $$
> 0\le soc(t)\le 16,\quad soc(T)=0,\quad \sum_t d(t)\le 32
> $$

---

## Step 0：目标统一成“最小化能量”
把常数项 $\sum_t p(t)w(t)$ 去掉（对决策无影响），等价变成：
- 最大化：$\sum_t p(t)\,(-c(t)+d(t))$
- 最小化：
$$
E=\sum_t p(t)\,[c(t)-d(t)]
$$

后续我们都用“最小化 $E$”的语言（QUBO/Ising 的标准形式）。

---

## Step 1：动作离散化（用 a(t) 决定充放电）
### 近似内容
用离散动作 $a(t)$ 直接决定 $(c(t),d(t))$，不再显式优化连续的 $c(t), d(t), y(t)$：

$$
a(t)\in\mathcal{A},\qquad (c(t),d(t)) = f(a(t))
$$

一个典型动作集合（你之前用过）：
- $a=0$: idle $(c,d)=(0,0)$  
- $a=1$: 半充 $(2.5,0)$  
- $a=2$: 满充 $(5,0)$  
- $a=3$: 半放 $(0,2)$  
- $a=4$: 满放 $(0,4)$  

### 近似后的约束/结构
- 互斥充放（原来的 $y(t)$）被**结构性消除**：动作表本身保证同一时刻不同时充放
- 功率上下界（5、4）也被表内固定
- 保留仍需处理的硬约束：SoC 动力学、SoC 边界、终端 SoC、总放电量

### 目标能量（按动作）
$$
E=\sum_t p(t)\,[c(a_t)-d(a_t)]
$$

### 比特粗算
若直接逐小时做动作（24个时刻），动作数 $|\mathcal{A}|=5$：
- one-hot：每时刻 5 bit → $24\times5=120$ bit（不可行）
- 二进制编码：$\lceil\log_2 5\rceil=3$ bit/时刻 → $24\times3=72$ bit（仍偏大）

---

## Step 2：时间块化（2小时 = 1 block）
### 近似内容
把 24 个时刻合并成 $B=12$ 个 block（每 block 2 小时），在 block 内动作恒定：
$$
a_b,\quad b=1,\dots,12
$$
对应的（可选）聚合价格与风：
$$
p_b=\frac{1}{2}(p_{2b-1}+p_{2b}),\quad w_b=\frac{1}{2}(w_{2b-1}+w_{2b})
$$

### 近似后的约束
SoC 动力学用 block 形式（仍保持同样结构）：
$$
soc_b = soc_{b-1} + \eta\,c(a_b) - d(a_b)
$$

### 目标能量
$$
E=\sum_{b=1}^{12} p_b\,[c(a_b)-d(a_b)]
$$

### 比特粗算
- 动作 3 bit/block → $12\times3=36$ bit ✅（这就是你“36比特方案”的核心）

---

## Step 3：SoC 粗量化（离散状态集）
你提出两种粒度（这里按“block 后”那种更一致）：

### 近似内容
block 后 SoC 只允许：
$$
soc_b \in \{0,4,8,16\}
$$

### 约束表达（离散状态机观点）
SoC 递推不再是连续值，而是“状态转移”：
$$
soc_b = T(soc_{b-1}, a_b)
$$
其中 $T$ 把 $(soc_{b-1}, a_b)$ 映射到最近/合法的离散 SoC（你可以选择：四舍五入、截断、或禁止非法转移）。

> 备注：如果你希望“硬禁止”，就对非法转移加大惩罚；如果希望“近似允许”，就用 rounding 近似转移。

### 比特粗算
- one-hot（4态）：每 block 4 bit → $12\times4=48$ bit
- 二进制（2 bit 表示 4 态）：每 block 2 bit → $12\times2=24$ bit

> 你后来采用的是“2bit SoC + 用深度/惩罚处理非法”，属于典型 **深度换比特**。

---

## Step 4：价格分裂（高价只放、低价只充）——消除互斥/降低耦合
### 近似内容
把 block 按价格排序，分成两半：
- 高价集合 $\mathcal{H}$：仅允许放电动作 $a\in\{3,4\}$
- 低价集合 $\mathcal{L}$：仅允许充电动作 $a\in\{1,2\}$（也可允许 idle）

这样每个 block 的动作集合变小，且“充放二选一”的冲突不再出现。

### 约束表达（动作域约束）
$$
a_b \in \mathcal{A}_b=
\begin{cases}
\{3,4\} & b\in\mathcal{H}\\
\{0,1,2\} & b\in\mathcal{L}
\end{cases}
$$

实现方式（QUBO/Ising 常用）：
- 对不允许的动作编码加惩罚项：
$$
E_{\text{domain}}=\lambda_{\text{dom}}\sum_b \mathbf{1}[a_b\notin\mathcal{A}_b]
$$

### 比特粗算
- 若高价 block 只需区分 {半放,满放}：1 bit/block
- 低价 block 区分 {idle,半充,满充}：2 bit/block（或用 2 bit 冗余编码）
整体变量数进一步下降（但你也可以继续用统一 3 bit 编码，只是多了非法惩罚）。

---

## Step 5：固定极值点（最高两点必放、最低两点必充）
### 近似内容
选定：
- 两个最高价 block：动作固定为放电（比如满放 $a=4$）
- 两个最低价 block：动作固定为充电（比如满充 $a=2$）

### 约束表达（变量冻结）
对固定 block，直接从优化变量中移除；等价于加无穷大惩罚：
$$
E_{\text{fix}}=\lambda_{\text{fix}}\sum_{b\in\mathcal{F}} \mathbf{1}[a_b\ne a_b^\star]
$$

### 比特粗算
每冻结一个 block，动作比特减少（例如 3 bit/block 直接少 3 个）。

---

## Step 6：SoC 编码从 one-hot → 2bit（用深度/惩罚处理 ancilla）
### 近似内容
用 2bit 表示 4 态：
- 例如 $(s_{b,1},s_{b,0})$ 表示 0/4/8/16（线性映射或查表映射）

若出现“编码态但不对应物理 SoC”（比如你以后扩展到 3bit/更多态），用惩罚项处理非法编码：
$$
E_{\text{soc-enc}}=\lambda_{\text{enc}}\sum_b \mathbf{1}[\text{soc code illegal}]
$$

### 比特粗算
SoC 寄存器：$12\times2=24$ bit（若你把 SoC 也作为变量进 QUBO 的话）

# 最终：给出一个“可落地的 Ising 模型”（面向 36 bit 的动作-only 版本）
> 你之前讨论的主线是：**用 36 bit 表示动作（12 block × 3 bit），SoC 约束用惩罚/后处理**。  
> 下面给的是这一版的最终 Ising。

---

## 1) 二进制变量定义（动作编码）
每个 block $b$ 用 3 个比特：
$$
x_{b,0},x_{b,1},x_{b,2}\in\{0,1\}
$$
动作编码：
$$
a_b = x_{b,0}+2x_{b,1}+4x_{b,2}\in\{0,\dots,7\}
$$
合法动作集合（根据 Step 1）：
$$
\mathcal{A}=\{0,1,2,3,4\}
$$

---

## 2) QUBO 能量（目标 + 惩罚）
我们最小化：
$$
E_{\text{QUBO}}(x)=E_{\text{obj}}(x)+E_{\text{inv}}(x)+E_{\text{soc}}(x)+E_{\Sigma d}(x)+E_{\text{domain}}(x)+E_{\text{fix}}(x)
$$

### (a) 目标项（真实优化充放电的核心）
对每个 block，用动作查表得到 $c(a_b),d(a_b)$：
$$
E_{\text{obj}}=\sum_{b=1}^{12} p_b\,[c(a_b)-d(a_b)]
$$
> 这一步在电路里对应 phase-separator（对每个 3bit 模式加相位）。

### (b) 非法动作惩罚（101/110/111）
$$
E_{\text{inv}}=\lambda_{\text{inv}}\sum_{b}\mathbf{1}[a_b\in\{5,6,7\}]
$$

### (c) SoC 动力学/边界/终端（软惩罚版本）
用动作序列在经典侧“计算”：
$$
soc_b = soc_{b-1}+\eta c(a_b)-d(a_b)
$$
惩罚：
$$
E_{\text{soc}}=\lambda_{\text{soc}}\left(
\sum_b \mathbf{1}[soc_b\notin[0,16]]
+\mathbf{1}[soc_{12}\ne 0]
\right)
$$

### (d) 总放电上限
$$
E_{\Sigma d}=\lambda_{d}\ \mathbf{1}\left[\sum_{b} d(a_b)>32\right]
$$

### (e) 价格分裂动作域（可选）
$$
E_{\text{domain}}=\lambda_{\text{dom}}\sum_b \mathbf{1}[a_b\notin\mathcal{A}_b]
$$

### (f) 固定极值点（可选）
$$
E_{\text{fix}}=\lambda_{\text{fix}}\sum_{b\in\mathcal{F}} \mathbf{1}[a_b\ne a_b^\star]
$$

> 说明：上面这些 $\mathbf{1}[\cdot]$ 统称“指示函数惩罚”。  
> 在 QUBO/Ising 里可以通过真值表展开成多项式（对 3bit 的局部条件尤其容易），或直接用量子线路的对角相位门实现。

---

## 3) 从 QUBO 到 Ising（标准变换）
若你最终得到标准 QUBO：
$$
E_{\text{QUBO}}(x)=\sum_i q_i x_i+\sum_{i<j} Q_{ij}x_ix_j + \text{const}
$$

定义 Ising 自旋变量：
$$
z_i\in\{-1,+1\},\qquad x_i=\frac{1-z_i}{2}
$$

则得到 Ising 形式：
$$
H(z)=\sum_i h_i z_i+\sum_{i<j} J_{ij} z_i z_j + \text{const}
$$
其中系数由代换得到（给你一个可直接用的系数关系）：
- 线性项：
$$
h_i = -\frac{q_i}{2}-\frac{1}{4}\sum_{j\ne i}Q_{ij}
$$
- 耦合项：
$$
J_{ij}=\frac{Q_{ij}}{4}
$$
（常数项可忽略，因为不影响最优解。）

# 比特预算（粗略）
下面给“你这条路线”最核心的几个版本对比：

## A) 动作-only（你 36 比特主线）
- 12 block × 3 bit = **36 qubits**
- SoC 不作为量子变量（经典评估 + 惩罚/后处理）
- 最适合上真实 54-qubit 机器：量子只做“动作组合搜索”

## B) 动作 + SoC 2bit（如果你想把 SoC 也放进 QUBO）
- 动作：36
- SoC：12 block × 2 bit = 24
- 合计：**60 qubits**（已经超过 54，除非减少 block 或冻结更多点）

## C) 动作域分裂 + 冻结极值点（进一步减比特）
- 高价 half：每 block 1 bit
- 低价 half：每 block 2 bit
- 再冻结 4 个 block（2 高 2 低）
- 总 qubits 常可压到 **20–30** 范围（很适合 NISQ）

# 最终你“最后的 Ising 模型”长什么样（总结一句）
你最后跑在量子机上的形式就是：

- 变量：$z_i$（共 36 个，对应 12 个 block 的动作比特）

$$
H(z)=\sum_i h_i z_i + \sum_{i<j} J_{ij} z_i z_j + \text{const}

- 其中 $h_i,J_{ij}$ 来自：
  1) 每个 block 的动作收益（$p_b(c-d)$ 的对角能量）  
  2) 非法动作惩罚
